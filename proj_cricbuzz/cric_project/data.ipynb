```python
def extract_matches(obj, rows):
    if isinstance(obj, dict):
        if "matchInfo" in obj:
            info = obj["matchInfo"]
            rows.append({
                "match_id": info.get("matchId"),
                "team1": info.get("team1", {}).get("teamName"),
                "team2": info.get("team2", {}).get("teamName"),
                "venue": info.get("venueInfo", {}).get("ground"),
                "city": info.get("venueInfo", {}).get("city"),
                "date": info.get("startDate")
            })
        # Recurse into dict values
        for v in obj.values():
            extract_matches(v, rows)
    elif isinstance(obj, list):
        for item in obj:
            extract_matches(item, rows)

rows = []
extract_matches(data, rows)

df_matchList = pd.DataFrame(rows)
df_matchList
```

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine(f"postgresql://postgres:DEVI@localhost:5432/cricbuzz_db")

# Collect unique player names from all tables
sources = ["matchlean", "matchstat", "overs"]
unique_names = set()

for src in sources:
    df = pd.read_sql(f"SELECT * FROM {src}", engine)
    for col in df.columns:
        if any(k in col.lower() for k in ["batsman", "bowler", "batter", "name"]):
            unique_names.update(df[col].dropna().astype(str).str.strip().unique())

# Convert to DataFrame
df_players = pd.DataFrame(sorted(unique_names), columns=["name"])

# Merge with existing player table (avoid duplicates)
existing = pd.read_sql("SELECT name FROM playerslist", engine)
existing_names = set(existing["name"].str.strip())

new_players = df_players[~df_players["name"].isin(existing_names)]

# Insert new players
new_players.to_sql("playerslist", engine, if_exists="append", index=False)
print(f" Added {len(new_players)} new unique players to playerslist table.")

```
